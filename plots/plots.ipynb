# Solution for various D

```python
import numpy as np
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import matplotlib.colors as mcolors
from matplotlib.colors import PowerNorm, BoundaryNorm, ListedColormap
from matplotlib import ticker
from scipy.optimize import curve_fit
from pylab import *
from scipy import ndimage
from matplotlib.colors import LinearSegmentedColormap
from functools import reduce
import operator
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams.update({
    "text.usetex": True,
    'font.size': 16,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'axes.labelsize': 12
})

# Create grid
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)

# Single deterministic function
def f(X, Y):
    return np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(12, 9))

for i in range(12):
    ax = fig.add_subplot(4, 3, i + 1, projection='3d')
    Z = f(X, Y)
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0, antialiased=False)
    ax.tick_params(direction='in', pad=-2)
    ax.set_xlabel(r"$\tau$",labelpad=-2,fontsize=12)
    ax.set_ylabel(r"$x$",labelpad=-2,fontsize=12)
    ax.set_zlabel(r"$z$",labelpad=-8,fontsize=12)
    ax.grid(False)
    ax.minorticks_on()

# Column labels
column_labels = [r"$D=3.1$", r"$D=4$", r"$D=5.5$"]

# x positions centered over the 3 columns
x_positions = [0.32, 0.5, 0.68]

y_position = 0.87  # vertical position near top of figure

for x, label in zip(x_positions, column_labels):
    fig.text(x, y_position, label, ha='center', va='bottom')

plt.subplots_adjust(wspace=-0.8, hspace=0.6)
plt.tight_layout(rect=[0,0, 1, 0.9])  # leave space at top for labels
plt.show()

plt.tight_layout()
plt.show()
```
