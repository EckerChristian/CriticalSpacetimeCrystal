```python
import numpy as np
import matplotlib.pyplot as plt

# LaTeX font (as you used)
plt.rcParams.update({
    "text.usetex": True,
    "font.size": 16
})

# Load data: columns assumed [D, value]
Delta = np.loadtxt("DeltaList.dat")
gamma = np.loadtxt("GammaList.dat")

# Sort by D
Delta = Delta[np.argsort(Delta[:, 0])]
gamma = gamma[np.argsort(gamma[:, 0])]

# Match by common D values (robust to float representation)
D_common = np.intersect1d(Delta[:, 0], gamma[:, 0])

# If your D values are not exactly identical in both files, use isclose matching instead:
def match_values_by_D(D_ref, arr, atol=1e-12, rtol=0.0):
    vals = np.empty_like(D_ref, dtype=float)
    for i, D in enumerate(D_ref):
        m = np.isclose(arr[:, 0], D, atol=atol, rtol=rtol)
        if not np.any(m):
            vals[i] = np.nan
        else:
            vals[i] = arr[m, 1][0]
    return vals

# Try exact intersection first; if empty, fall back to "use Delta's D grid"
if D_common.size > 0:
    Delta_c = match_values_by_D(D_common, Delta)
    gamma_c = match_values_by_D(D_common, gamma)
    D_plot = D_common
else:
    D_plot = Delta[:, 0]
    Delta_c = Delta[:, 1]
    gamma_c = match_values_by_D(D_plot, gamma)

# Ratio (mask invalid / missing)
ratio = Delta_c / gamma_c
mask = np.isfinite(ratio)

# Plot
fig, ax = plt.subplots(facecolor="white", dpi=100, figsize=(6, 4))
ax.tick_params(which="major", direction="in", top=True, right=True)
ax.tick_params(which="minor", direction="in", top=True, right=True)
ax.minorticks_on()

ax.set_xlabel(r"${\rm Dimension}~D$", fontsize=16)
ax.set_ylabel(r"$\Delta/(2\gamma)$", fontsize=16)

ax.plot(D_plot[mask], ratio[mask]/2, color="black", linewidth=2)

# Optional: mark D=4 if present
m4 = np.isclose(D_plot, 4.0, atol=1e-12, rtol=0.0) & mask
if np.any(m4):
    ax.plot(D_plot[m4][0], ratio[m4][0]/2, "o", color="black")
    ax.text(4.0 , 5, rf"$D=4,\ \Delta/(2\gamma)={ratio[m4][0]/2:.6f}$", fontsize=16)

plt.savefig("Delta_over_gamma.pdf", dpi=100, bbox_inches="tight")
```

# large-D power law

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.interpolate import UnivariateSpline

# LaTeX font
plt.rcParams.update({
    "text.usetex": True,
    "font.size": 16
})

def weighted_spline_samples(keys, values, errors, n_samples=300, k=2, s=None):
    """
    keys   : x-values (list/array)
    values : y-values (list/array)
    errors : 1 y-errors (list/array), used as weights (smaller error => higher weight)
    n_samples: number of equally spaced x-samples to return
    k      : spline degree (k=2 quadratic)
    s      : smoothing factor. If None, use ~N (good starting point when errors are 1).
    Returns: xx (equidistant), yy (spline evaluated), spline object
    """
    x = np.asarray(keys, dtype=float)
    y = np.asarray(values, dtype=float)
    sig = np.asarray(errors, dtype=float)

    # Basic cleaning
    m = np.isfinite(x) & np.isfinite(y) & np.isfinite(sig) & (sig > 0)
    x, y, sig = x[m], y[m], sig[m]
    if x.size < (k + 1):
        raise ValueError(f"Need at least {k+1} valid points for a degree-{k} spline; got {x.size}.")

    # Sort by x (required)
    idx = np.argsort(x)
    x, y, sig = x[idx], y[idx], sig[idx]

    # Handle duplicate x-values by averaging (weighted by 1/sigma^2)
    uniq_x, inv = np.unique(x, return_inverse=True)
    if uniq_x.size != x.size:
        w2 = 1.0 / (sig**2)
        y_acc = np.zeros_like(uniq_x, dtype=float)
        w_acc = np.zeros_like(uniq_x, dtype=float)

        for i, j in enumerate(inv):
            y_acc[j] += y[i] * w2[i]
            w_acc[j] += w2[i]

        y = y_acc / w_acc
        sig = 1.0 / np.sqrt(w_acc)
        x = uniq_x

    # Weights for UnivariateSpline: w = 1/sigma
    w = 1.0 / sig
    if s is None:
        s = len(x)

    spl = UnivariateSpline(x, y, w=w, s=s, k=k)

    xx = np.linspace(x.min(), x.max(), n_samples)
    yy = spl(xx)
    return xx, yy, spl


def power_law(x, a, b):
    return a * np.power(x, b)


# ---------------- Load data ----------------
deltadat = np.loadtxt("DeltaList.dat")
Delta = np.loadtxt("DeltaList.dat")

SO = np.loadtxt("DeltaSorkin.dat")
bland1 = np.loadtxt("DeltaBlandThesis.dat")
bland2 = np.loadtxt("DeltaBlandPaper1.dat")
bland3 = np.loadtxt("DeltaBlandPaper2.dat")

# ---------------- Dashed fit: power-law using the largest-D points from DeltaList ----------------
D_all = deltadat[:, 0]
Delta_all = deltadat[:, 1]

N_largest = 10  # you currently use 4; change back to 10 if desired
idxN = np.argsort(D_all)[-N_largest:]
DN = D_all[idxN]
DeltaN = Delta_all[idxN]

xN = 1.0 / DN
yN = DeltaN

mN = (xN > 0) & (yN > 0) & np.isfinite(xN) & np.isfinite(yN)
b0, loga0 = np.polyfit(np.log(xN[mN]), np.log(yN[mN]), 1)
p0 = (np.exp(loga0), b0)

popt, pcov = curve_fit(power_law, xN[mN], yN[mN], p0=p0, maxfev=20000)
a_fit, b_fit = popt
a_err, b_err = np.sqrt(np.diag(pcov))

print(f"Dashed fit uses D in [{DN.min():.6g}, {DN.max():.6g}] (largest {N_largest} from DeltaList.dat)")
print(f"a = {a_fit:.8g}  {a_err:.2g}")
print(f"b = {b_fit:.8g}  {b_err:.2g}")

# x-range for dashed curve (matches your previous intent)
xx = np.linspace(0.02, xN.max(), 300)
yy = power_law(xx, a_fit, b_fit)

# ---------------- WEIGHTED QUADRATIC SPLINE (k=2) USING ALL D>4 DATA ----------------

Dmin = 4.0

# this work (D>4)
m_this = deltadat[:, 0] > Dmin
x_spl = 1.0 / deltadat[m_this, 0]
y_spl = deltadat[m_this, 1]
e_spl = np.full_like(y_spl, 1.0e-5, dtype=float)  # effectively exact

# bland thesis (D>4)
m_b1 = bland1[:, 0] > Dmin
x_spl = np.concatenate([x_spl, 1.0 / bland1[m_b1, 0]])
y_spl = np.concatenate([y_spl, bland1[m_b1, 1]])
e_spl = np.concatenate([e_spl, bland1[m_b1, 2]])

# bland paper 1 (D>4)
m_b2 = bland2[:, 0] > Dmin
x_spl = np.concatenate([x_spl, 1.0 / bland2[m_b2, 0]])
y_spl = np.concatenate([y_spl, bland2[m_b2, 1]])
e_spl = np.concatenate([e_spl, bland2[m_b2, 2]])

# bland paper 2 / BK (D>4)
bland3_arr = np.atleast_2d(bland3)
m_b3 = bland3_arr[:, 0] > Dmin
x_spl = np.concatenate([x_spl, 1.0 / bland3_arr[m_b3, 0]])
y_spl = np.concatenate([y_spl, bland3_arr[m_b3, 1]])
e_spl = np.concatenate([e_spl, bland3_arr[m_b3, 2]])

# SO data (D>4)   NEW
m_so = SO[:, 0] > Dmin
x_spl = np.concatenate([x_spl, 1.0 / SO[m_so, 0]])
y_spl = np.concatenate([y_spl, SO[m_so, 1]])
e_spl = np.concatenate([e_spl, SO[m_so, 2]])

# build spline
xx_spl, yy_spl, spl = weighted_spline_samples(
    x_spl, y_spl, e_spl,
    n_samples=300,
    k=2,
    s=None
)

# ---------------- Dotted fit: power-law using the smallest 1/D end of the spline (extend to same xmin as dashed) ----------------
N_head = 10  # number of first spline samples to fit

x_head = xx_spl[:N_head]
y_head = yy_spl[:N_head]

# print D-range used for the dotted-line (small-1/D) fit
D_head_min = 1.0 / np.max(x_head)
D_head_max = 1.0 / np.min(x_head)

print(f"Dotted-line fit uses spline points in the range:")
print(f"    D  [{D_head_min:.6g}, {D_head_max:.6g}]")

m_head = (x_head > 0) & (y_head > 0) & np.isfinite(x_head) & np.isfinite(y_head)
b0_head, loga0_head = np.polyfit(np.log(x_head[m_head]), np.log(y_head[m_head]), 1)
p0_head = (np.exp(loga0_head), b0_head)

popt_head, pcov_head = curve_fit(power_law, x_head[m_head], y_head[m_head], p0=p0_head, maxfev=20000)
a_head, b_head = popt_head
a_head_err, b_head_err = np.sqrt(np.diag(pcov_head))

print(f"Dotted fit uses first {N_head} spline samples (smallest 1/D end).")
print(f"a = {a_head:.8g}  {a_head_err:.2g}")
print(f"b = {b_head:.8g}  {b_head_err:.2g}")

x_min_dashed = np.min(xx)          # extend dotted down to same xmin as dashed
x_max_dotted = np.max(x_head)      # keep its upper end at the end of the fitted spline head

xx_headfit = np.logspace(np.log10(x_min_dashed), np.log10(x_max_dotted), 400)
yy_headfit = power_law(xx_headfit, a_head, b_head)

# ---------------- Plot ----------------
fig, ax = plt.subplots(sharex=True, facecolor="white", dpi=100, figsize=(6, 4))
ax.set_xlim([0.06, 0.22])
ax.set_ylim([1.9, 3.6])
ax.tick_params(which="major", direction="in", top=True, right=True)
ax.tick_params(which="minor", direction="in", top=True, right=True)
ax.minorticks_on()
ax.set_xlabel(r"$1/D$", fontsize=16)
ax.set_ylabel(r"$\Delta$", fontsize=16)

# this work
ax.plot(1.0 / Delta[:, 0], Delta[:, 1],
        color="black", linestyle="-", linewidth=4, zorder=1, alpha=1,
        label=r"$\rm this~work$")

# error bars in literature (excluded from main legend; separate legend bottom right)
h_SO = ax.errorbar(1.0 / SO[:, 0], SO[:, 1], yerr=SO[:, 2],
                   fmt="o", markersize=3, capsize=6, zorder=2, linewidth=1,
                   label="_nolegend_")
h_b2 = ax.errorbar(1.0 / bland2[:, 0], bland2[:, 1], yerr=bland2[:, 2],
                   fmt="o", markersize=3, capsize=6, zorder=2, linewidth=1,
                   label="_nolegend_")
h_b3 = ax.errorbar(1.0 / bland3_arr[:, 0], bland3_arr[:, 1], yerr=bland3_arr[:, 2],
                   fmt="o", markersize=3, capsize=6, zorder=2, linewidth=1,
                   label="_nolegend_")

# dashed: large-D fit to this work
ax.plot(xx, yy, color="black", linestyle="--", linewidth=2, zorder=3,
        label=r"$\propto(1/D)^{0.37}$")

# spline
ax.plot(xx_spl, yy_spl, color="red", linestyle="-", linewidth=1,
        label=r"$\rm weighted~spline$")

# dotted: small-1/D spline-end fit, extended to xmin of dashed
ax.plot(xx_headfit, yy_headfit, color="black", linestyle=":", linewidth=2, zorder=4,
        label=r"$\propto(1/D)^{0.31}$")

# main legend (lines only)
leg1 = ax.legend(loc="upper left", fontsize=14, frameon=False)

# second legend for errorbar datasets
leg2 = ax.legend([h_SO, h_b2, h_b3],
                 [r"$\rm SO$", r"$\rm BPBKH$", r"$\rm BK$"],
                 loc="lower right", fontsize=14, frameon=False)

ax.add_artist(leg1)

plt.savefig("DeltaLargeD.pdf", dpi=100, bbox_inches="tight")
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# ---------------- LaTeX Font ----------------
plt.rcParams.update({
    "text.usetex": True,
    "font.size": 22
})

# ---------------- Load Data ----------------
with open("../complete_data/critcoll_data.json", "r") as f:
    data = json.load(f)

dims = ["3.05", "3.2", "3.4", "3.6", "3.8", "4.0", "4.2", "4.4", "4.6", "4.8", "5.0", "5.199", "5.5"]
vals = np.array(dims,dtype=float)

Delta = [data[dim]["Initial_Conditions"]["Delta"] for dim in dims]

fc = []
psic = []
psimp = []

for dim in dims:
    tauvals = np.linspace(0.,1.,data[dim]["Ntau"],endpoint=False)
    fcnow = np.array(list(zip(tauvals,data[dim]["Initial_Conditions"]["fc"])))
    fact = np.sqrt((float(dim)-2)**3 / 4)
    psimpvals = list(fact * np.array(data[dim]["Initial_Conditions"]["Up"]))
    psicnow = np.array(list(zip(tauvals,data[dim]["Initial_Conditions"]["psic"])))
    psimpnow = np.array(list(zip(tauvals,psimpvals)))
    fc.append(fcnow)
    psic.append(psicnow)
    psimp.append(psimpnow)

# ---------------- Color Parameter ----------------
vmin, vmax = vals.min(), vals.max()
norm_vals = (vals - vmin) / (vmax - vmin)
cmap = cm.get_cmap("coolwarm_r")

# =================================================
#  Helper: Curve Plotter
# =================================================
def plot_family(ax, arr, ylabel):
    """
    arr[i,:,0] = x
    arr[i,:,1] = y
    """
    for i in range(len(arr)):
        x = arr[i][:,0]
        y = arr[i][:,1]
        x_norm = (x - x.min()) / (x.max() - x.min())
        y_norm=y/y.max()
        if ax==axes[1]:
            y=y_norm
        ax.plot(x_norm, y, color=cmap(norm_vals[i]), linewidth=1)

    ax.set_xlim(0, 1)
    #ax.set_ylim(0, 1.1)
    ax.set_xlabel(r"$\tau/\Delta$")
    ax.set_ylabel(ylabel)

    # ticks in all directions
    ax.tick_params(direction="in", which="both", top=True, right=True)
    ax.minorticks_on()


# =================================================
#  Main Figure
# =================================================
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=False)

axes[0].set_ylim(-0.02, 0.82)
axes[1].set_ylim(-1.05, 1.04)
axes[2].set_ylim(-1.05, 1.04)

plot_family(axes[0], fc,   r"$f_c$")
plot_family(axes[1], psic, r"$\Psi_c/{\rm max}(\Psi_c)$")
plot_family(axes[2], psimp,   r"$\psi_{-p}$")

axes[0].set_ylabel(r"$f_c$", rotation=0, labelpad=40)
axes[0].yaxis.set_label_coords(0, 1.02)

axes[1].set_ylabel(r"$\Psi_c/{\rm max}(\Psi_c)$", rotation=0, labelpad=40)
axes[1].yaxis.set_label_coords(0, 1.02)

axes[2].set_ylabel(r"$\psi_{-p}$", rotation=0, labelpad=40)
axes[2].yaxis.set_label_coords(0, 1.02)

# =================================================
#  Dedicated Colorbar Axis (outside)
# =================================================
sm = cm.ScalarMappable(cmap=cmap)
sm.set_array(vals)

# Create a vertical bar to the right of last axis
cbar_ax = fig.add_axes([0.92, 0.1, 0.02, 0.70])  # left, bottom, width, height
cbar = fig.colorbar(sm, cax=cbar_ax, orientation="vertical")

cbar.set_label(r"${\rm Dimension}~D$")
cbar.ax.tick_params(direction="in")


plt.subplots_adjust(left=0, right=0.90, wspace=0.25, top=0.8)
plt.savefig("boundaryConditions.pdf", dpi=100, bbox_inches="tight")
plt.show()
```
